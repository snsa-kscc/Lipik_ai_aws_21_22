```python
import pandas as pd
import numpy as np
import os
from datetime import date

url = "https://raw.githubusercontent.com/deksa89/git-project-lipik/master/web%20scraping%20i%20environment/letovi.csv"

df = pd.read_csv(url)

months = {
    'Dec': 12,
    'Jan': 1,
    'Mar': 3,
    'Apr': 4
}

def get_propper_date(input_date):

    date_array = input_date.split(' ')
    day = int(date_array[1])
    month_word = date_array[2]
    month = months[month_word]
    year = 2020 if month == 12 else 2021
    return date(year, month, day)


df['formatted_departure'] = pd.Series(map(get_propper_date, df['Departure date:']))
df['formatted_arrival'] = pd.Series(map(get_propper_date, df['Date of Arrival:']))

df['dif'] = np.where(
    df['formatted_departure'] == df['formatted_arrival'], '1h', (df['formatted_arrival'] - df['formatted_departure']).dt.days)

df
```
